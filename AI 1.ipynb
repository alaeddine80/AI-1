```python
#Question 7
#Square root of [(2 * C * D)/H] 
import numpy as np
C=50
H=30
D=[]  
n = int(input())
for i in range(0, n):
    ele = int(input())
    D.append(ele) 
print(D)
for ele in D:
    A=np.sqrt([(2*C*ele)/H])
    print(int(A))
```
